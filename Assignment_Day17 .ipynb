```python
import numpy as np
import pandas as pd

os.chdir('D:\Dataset')
df = pandas.read_csv("mtcars.csv")
x=df['hp']
outliers=[]
def detect_outlier(data_1):
    
    threshold=2
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)
    print("The Mean",mean_1)
    print("The STD",std_1)
    
    for y in data_1:
        z_score= (y - mean_1)/std_1 
      
        if np.abs(z_score) > threshold:
              outliers.append(y)
            
    return outliers

outlier_datapoints = detect_outlier(df['hp'])
print("outlier Points",outlier_datapoints)
```
